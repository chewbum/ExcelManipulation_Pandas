# Task 2

```python
companies['SMA_5'] = companies.groupby('TICKER')['PRICECLOSE'].rolling(window=5, min_periods=5).mean().reset_index(0, drop=True)
```

# Task 3

```python
companies['DAILY_RETURN'] = companies.groupby('TICKER')['PRICECLOSE'].pct_change() * 100
```

# Task 4

```python
from datetime import timedelta
print(companies.columns)

def get_prev_date(row):
    curr = row['DATE']
    ticker = row['TICKER']
    prev = curr - timedelta(days = 30)
    # Iterate back until a trading day is found
    ticker_rows = companies[companies['TICKER'] == ticker]
    while not ticker_rows.empty and prev not in ticker_rows['DATE'].values:
        prev = prev + timedelta(days = 1)
    return prev if not ticker_rows.empty else curr

companies['DATE'] = pd.to_datetime(companies['DATE'], dayfirst = True) # Ensure 'DATE' column is in datetime format
companies['new_date'] = companies.apply(get_prev_date, axis = 1)


# Create a lookup dictionary for the close price for each ticker on each date
price_lookup = companies.set_index(['TICKER', 'DATE'])['PRICECLOSE'].to_dict()

def get_price_close(row, date_col):
    return price_lookup.get((row['TICKER'], row[date_col]), None)

companies['price_close_new_date'] = companies.apply(lambda row: get_price_close(row, 'new_date'), axis=1)
companies['MONTHLY_RETURN'] = (companies['PRICECLOSE'] - companies['price_close_new_date']) / companies['price_close_new_date'] * 100

companies.head(32)
```

# Question 8

```python
# Ensure 'DATE' is in datetime format
companies['DATE'] = pd.to_datetime(companies['DATE'], dayfirst = True) 

# Filter DataFrame for 'Financials' sector and specified date range
filtered = companies[
    (companies['SECTOR'] == 'Financials') &
    (companies['DATE'] >= '2021-01-01') &
    (companies['DATE'] <= '2021-06-30')
]

# Find the ticker of the stock with the lowest 'DAILY_RETURN', return the index of min
ticker_with_lowest_return = filtered.loc[filtered['DAILY_RETURN'].idxmin()]['TICKER']

print(ticker_with_lowest_return)
```

# Question 9

```python
# Filter DataFrame for "Information Technology" sector
it_sector = companies[companies['SECTOR'] == 'Information Technology']

# Find the index of the row with the highest MONTHLY_RETURN
highest_return_index = it_sector['MONTHLY_RETURN'].idxmax()

# Get the TICKER of the stock with the highest MONTHLY_RETURN
ticker = it_sector.loc[highest_return_index, 'TICKER']
print(ticker)
```

# Question 10

```python
companies['DATE'] = pd.to_datetime(companies['DATE'], dayfirst = True) 

# Filter DataFrame for the date range 1 Jan 2021 - 31 December 2021 inclusive
#returns a series of boolean values (same length as the number of rows in the df)
mask = (companies['DATE'] >= '2021-01-01') & (companies['DATE'] <= '2021-12-31')
#loc is used for indexing by label but it can also accept a boolean mask as input, selecting true
filtered_companies = companies.loc[mask]

# Group by 'SECTOR' and calculate average 'MONTHLY_RETURN'
sector_performance = filtered_companies.groupby('SECTOR')['MONTHLY_RETURN'].mean()

# Find the sector with the highest average 'MONTHLY_RETURN'
best_sector = sector_performance.idxmax()

print(best_sector)
```
